```python
#im gonna start by defining some variables that are useable in the apple api.

#first is term this will always be required since its kinda useless to not search with it.

#try changing this to any show you want. it will make a new json document with a name of based on the search term, just use + instead of spaces.
term = 'Curious+George' #input('what term do you want to search')
#next will be required for the assignment and thats media, it defines what type of media will be used, for this im gonna use a list so when I make my personal project I can use this list to lessen the chances of user error
medialist = ['movie', 'podcast', 'music', 'musicVideo', 'audiobook', 'shortFilm', 'tvShow', 'software', 'ebook', 'all']
#just gonna hard code this for the lab
media = 'tvShow'


#later on when I do my personal project im gonna make it so this can have optional variables but for now this is what we need
api_url = f'https://itunes.apple.com/search?term={term}&media={media}'

#print(api_url)
```

getting the request and json formatted content

```python
r = requests.get(api_url)
r_json = json.loads(r.content)
r_json
```

lets see how this looks without getting deeper into it 

```python
results = pd.json_normalize(r_json, record_path =["results"])
#we definetely need to gett deeper, added the results portion and now it looks good just kind of... messy. gonna do the sort next.
```

```python
results
```

sorting the dataframe by collection name and track number

```python
sorted_results = results.sort_values(by=['collectionName', 'trackNumber'])
sorted_results
#there are a few missing seasons and episodes :(


```

Create a JSON file that displays nested JSON: group by the artist & collection name. At the nested level, show the trackID, track number, track name, and short description. Make sure the JSON is formatted nicely in the file.

Basically to do this im gonna copy what you did with the employee dataframe but im writting in the columns from apple instead

```python
nestList = ["trackId", "trackNumber", "trackName", "shortDescription"]

j = (sorted_results.sort_values(by=["artistName", "collectionName"])
                   .groupby(["artistName", "collectionName"]) # Group the dataframe
                   .apply(lambda x: x[nestList].to_dict("records")) # Make the nested items be a dictionary
                   .reset_index() # Reset the index
                   .rename(columns={0: "nestedItems"}) # Rename the new column
                   .to_json(orient="records")) # Spit it out as JSON

cjson = json.dumps(json.loads(j), indent = 3) # make it pretty
cjson
#I decided to do it by the artistName (which is just the fictional artist of Curious George lmao) even tho id argue that the orignal book author is the real artist. Didnt do the artistID
#also gotta appreciate apple using camelCase with their columnNames
```

now to save it to a json file

```python
with open(f"{term}Apple{media}.json", "w") as file:
    file.write(cjson)
```
